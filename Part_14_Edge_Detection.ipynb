```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cats_img = cv2.imread('resources/Photos/cats.jpg')
cats_img = cv2.cvtColor(cats_img, cv2.COLOR_BGR2GRAY)
plt.imshow(cats_img, cmap='gray')
```

```python
# Laplaction
lap_img = cv2.Laplacian(cats_img, cv2.CV_64F)
lap_img = np.uint8(np.absolute(lap_img))
plt.figure(figsize=(12, 16))
plt.imshow(lap_img, cmap='gray')
```

```python
# Sobel
sobel_x_img = cv2.Sobel(cats_img, cv2.CV_64F, 1, 0)
sobel_y_img = cv2.Sobel(cats_img, cv2.CV_64F, 0, 1)
plt.figure(figsize=(22, 26))
plt.subplot(1, 2, 1)
plt.imshow(sobel_x_img, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(sobel_y_img, cmap='gray')
```
