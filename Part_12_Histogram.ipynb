```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston_img = cv2.imread('resources/Photos/park.jpg')
boston_img = cv2.cvtColor(boston_img, cv2.COLOR_BGR2RGB)
plt.imshow(boston_img)
```

```python
gray_boston_img = cv2.cvtColor(boston_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_boston_img, cmap='gray')
```

```python
gray_hist = cv2.calcHist([gray_boston_img], [0], None, [256], [0, 256])
plt.plot(gray_hist, color='gray')
```

```python
plt.figure(figsize=(22, 8))
for idx, color in enumerate('rgb'):
    hist = cv2.calcHist([boston_img], [idx], None, [256], [0, 256])
    plt.subplot(1, 3, idx+1)
    plt.title(f'Color channel: {color.upper()}')
    plt.plot(hist, color=color)
plt.show()
```
