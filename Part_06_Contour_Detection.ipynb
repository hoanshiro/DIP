```python
cat_img = cv2.imread('resources/Photos/cats.jpg')
cat_img = cat_img[...,::-1]
plt.imshow(cat_img)
```

```python
gray_cat_img = cv2.cvtColor(cat_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_cat_img, cmap='gray')
```

```python
# detect edges
canny_cat_img = cv2.Canny(cat_img, 125, 175)
plt.imshow(canny_cat_img, cmap='gray')
```

```python
contours_cat_img, hierarchies = cv2.findContours(canny_cat_img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
print(f"{len(contours_cat_img)} contours found")
```

```python
ret, thresh = cv2.threshold(gray_cat_img, 125, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(12, 16))
plt.imshow(thresh, cmap='gray')
```

```python
thresh_cat_img, hierarchies = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
print(f"{len(thresh_cat_img)} contours found")
```

```python
plt.figure(figsize=(12, 16))
white_backgroud = np.full(cat_img.shape, 255, dtype='uint8')
cv2.drawContours(white_backgroud, contours_cat_img, -1, (0, 100, 255), 2)
plt.imshow(white_backgroud)
```
