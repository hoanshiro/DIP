```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cats_img = cv2.imread('resources/Photos/cats.jpg')
cats_img = cv2.cvtColor(cats_img, cv2.COLOR_BGR2RGB)
plt.imshow(cats_img)
```

```python
gray_cats_img = cv2.cvtColor(cats_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_cats_img, cmap='gray')
```

```python
threshold, thresh_img = cv2.threshold(gray_cats_img, 150, 255, cv2.THRESH_BINARY)
plt.imshow(thresh_img, cmap='gray')
```

```python
threshold, thresh_inv_img = cv2.threshold(gray_cats_img, 150, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh_inv_img, cmap='gray')
```

```python
adaptive_thresh_img = cv2.adaptiveThreshold(gray_cats_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                           cv2.THRESH_BINARY, 11, 3)
plt.figure(figsize=(12, 16))
plt.imshow(adaptive_thresh_img, cmap='gray')
```
