```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
black_matrix = np.zeros((1000, 1000, 3), dtype='uint8')
plt.imshow(black_matrix)
```

```python
white_matrix = np.full((500, 800, 3), 255, dtype='uint8')
plt.imshow(white_matrix)
```

```python
green_matrix = black_matrix.copy()
green_matrix[:] = 0, 255, 0
plt.imshow(green_matrix)
```

```python
green_matrix.shape
```

```python
np.random.seed(42)
rand_matrix = np.random.randint(0, 255, (1080, 1920, 3), dtype='uint8')
plt.imshow(rand_matrix)
```

```python
draw_backgroud = rand_matrix.copy()
start_point = (255, 255)
end_point = (500, 500)
color = (0, 255, 0)
thickness = 8
# thickness = cv.FILLED or -1
cv2.rectangle(draw_backgroud, start_point, end_point, color, thickness=thickness)

center = np.multiply(draw_backgroud.shape[:2], 0.5).astype(int)[::-1]
cv2.circle(draw_backgroud, center, 100, color, thickness=thickness)

cv2.line(draw_backgroud, start_point ,end_point, color, thickness=thickness)
plt.imshow(draw_backgroud)
```

```python
text_background = white_matrix.copy()
cv2.putText(text_background, "Hello, World", (225, 225),
            cv.FONT_HERSHEY_TRIPLEX, 2, (255, 0, 0), 2)
plt.imshow(text_background)
```
