## 1. Resize Image

```python
cat_large_img = cv.imread('resources/Photos/cats 2.jpg')
cat_large_img = cv.cvtColor(cat_large_img, cv.COLOR_BGR2RGB)
plt.imshow(cat_large_img)
```

```python
cat_large_img.shape
```

```python
def rescaleFrame(frame, ratio=0.75):
    # works for all image, video, live
    height, width = np.multiply(cat_large_img.shape[:2], ratio).astype(int)
    dimensions = (width, height)
    rescaled_img = cv.resize(frame, dimensions, interpolation=cv.INTER_AREA)
    return rescaled_img
```

```python
rescaled_img = rescaleFrame(cat_large_img, ratio=0.5)
plt.imshow(rescaled_img)
```

```python
def changeResolution(capture, width, height):
    # only for Live
    capture.set(3, width) # 3 for image width
    capture.set(4, height) # 4 for image height
```
