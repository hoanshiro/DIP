```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
group_img = cv2.imread('resources/Photos/group 1.jpg')
group_img = cv2.cvtColor(group_img, cv2.COLOR_BGR2RGB)
gray_group_img = cv2.cvtColor(group_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_group_img, cmap='gray')
```

```python
haar_cascade = cv2.CascadeClassifier('resources/haarcascade_face.xml')
```

```python
faces_rect = haar_cascade.detectMultiScale(gray_group_img, scaleFactor=1.1, minNeighbors=3)
print(f"{len(faces_rect)} faces detected!")
```

```python
grop_clone_img = group_img.copy()
for (x, y, w, h) in faces_rect:
    cv2.rectangle(grop_clone_img, (x, y), (x+w, y+h), (0, 255, 0), thickness=2)
plt.figure(figsize=(12, 16))
plt.imshow(grop_clone_img)
```
