```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cats_img = cv2.imread('resources/Photos/cats.jpg')
cats_img = cv2.cvtColor(cats_img, cv2.COLOR_BGR2RGB)
plt.imshow(cats_img)
```

```python
white_img = np.full(cats_img.shape[:2], 0, dtype='uint8')
mask = cv2.circle(white_img, (cats_img.shape[1]//2, cats_img.shape[0]//2), 100, 255, -1)
masked_img = cv2.bitwise_and(cats_img, cats_img, mask=mask)
plt.imshow(masked_img)
```
