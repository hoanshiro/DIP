```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston_img = cv2.imread('resources/Photos/park.jpg')
```

```python
# convert BGR to GrayScale
gray_boston_img = cv2.cvtColor(boston_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_boston_img, cmap='gray')
```

```python
# convert BGR to HSV (Hue Saturation Value)
hsv_boston_img = cv2.cvtColor(boston_img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_boston_img[...,::-1])
```

```python
# convert BGR to HSV (Hue Saturation Value)
lab_boston_img = cv2.cvtColor(boston_img, cv2.COLOR_BGR2LAB)
plt.imshow(lab_boston_img[...,::-1])
```
