```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston_img = cv2.imread('resources/Photos/park.jpg')
boston_img = cv2.cvtColor(boston_img, cv2.COLOR_BGR2RGB)
plt.imshow(boston_img)
```

```python
list_channel_img = cv2.split(boston_img)
print("RGB")
plt.figure(figsize=(32, 36))
for idx, channel_img in enumerate(list_channel_img):
    plt.subplot(1, 3, idx+1)
    plt.imshow(channel_img, cmap='gray')
```

```python
merged_img = cv2.merge(list_channel_img)
plt.imshow(merged_img)
```
