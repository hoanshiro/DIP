```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston_img = cv2.imread('resources/Photos/park.jpg')
boston_img = boston_img[:,:,::-1]
plt.imshow(boston_img)
```

```python
def translate(img, x, y):
    '''
    parametrs:
    ----------
    x: positive -> shift image to right x pixels
       negative -> shift image to left x pixels
    y: positive -> shift image to down y pixels
       negative -> shift image to up y pixels
    '''
    transMat = np.float32([[1, 0, x], [0, 1, y]])
    dimensions = (img.shape[1], img.shape[0])
    return cv2.warpAffine(img, transMat, dimensions)
```

```python
plt.figure(figsize=(18, 22))
tranlated_park_img = translate(boston_img, 100, 100)

plt.subplot(1, 2, 1)
plt.imshow(tranlated_park_img)
tranlated_park_img = translate(boston_img, -100, -100)

plt.subplot(1, 2, 2)
plt.imshow(tranlated_park_img)
plt.plot()
```

```python
def rotate(img, angle, rotPoint=None):
    height, width = img.shape[:2]
    if rotPoint is None:
        rotPoint= (width//2, height//2)
    rotMat = cv2.getRotationMatrix2D(rotPoint, angle, 1.0)
    dimensions = (width, height)
    return cv2.warpAffine(img, rotMat, dimensions)
```

```python
plt.figure(figsize=(16, 22))

rotated_park_img = rotate(boston_img, 45, None)
plt.subplot(1, 3, 1)
plt.imshow(rotated_park_img)

rotated_park_img = rotate(boston_img, -5, (0, 0))
plt.subplot(1, 3, 2)
plt.imshow(rotated_park_img)

rotated_park_img = rotate(boston_img, 180)
plt.subplot(1, 3, 3)
plt.imshow(rotated_park_img)

plt.plot()
```

```python
plt.figure(figsize=(26, 30))
resized_img = cv2.resize(boston_img, (1000, 1000), interpolation=cv2.INTER_CUBIC)
plt.subplot(1, 2, 1)
plt.imshow(boston_img)

plt.subplot(1, 2, 2)
plt.imshow(resized_img)
```
