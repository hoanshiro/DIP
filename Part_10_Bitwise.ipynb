```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
white_backgroud = np.full((400, 400), 255, dtype='uint8')
retangle = cv2.rectangle(white_backgroud.copy(), (30, 30), (370, 370), 0, -1)
circle = cv2.circle(white_backgroud.copy(), (200, 200), 200, 0, -1)
```

```python
plt.figure(figsize=(32, 36))
for idx, img in enumerate([white_backgroud, retangle, circle]):
    plt.subplot(1, 3, idx+1)
    plt.imshow(img, cmap='gray')
```

```python
bitwise_and_img = cv2.bitwise_and(retangle, circle)
plt.imshow(bitwise_and_img, cmap='gray')
```

```python
bitwise_or_img = cv2.bitwise_or(retangle, circle)
plt.imshow(bitwise_or_img, cmap='gray')
```
